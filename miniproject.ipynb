```python
import pandas as pd
from datetime import datetime

df = pd.read_excel("Hospital  Health Care Management Data set.xlsx")

# Convert 'Date' to datetime
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Clean 'Status' column: lowercase + remove extra spaces
df['Status'] = df['Status'].astype(str).str.strip().str.lower()

total_discharges = df[df['Status'].str.contains("discharge")].shape[0]
print("Total Discharges:", total_discharges)

# Example: October 2007
selected_month = 8
selected_year = 2007

mtd_discharges = df[
    (df['Date'].dt.month == selected_month) &
    (df['Date'].dt.year == selected_year) &
    (df['Status'].str.contains("discharge"))
].shape[0]

print(f"MTD Discharges (Manual: {selected_month}/{selected_year}):", mtd_discharges)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel("Hospital  Health Care Management Data set.xlsx")

# Ensure AdmissionDate is in datetime format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

monthly_admissions = (
    df.groupby(df['Date'].dt.to_period('M'))
      .size()
      .reset_index(name='Admissions')
)

# Convert Period to Timestamp for plotting
monthly_admissions['Date'] = monthly_admissions['Date'].dt.to_timestamp()
plt.figure(figsize=(12, 6))

plt.fill_between(monthly_admissions['Date'],
                 monthly_admissions['Admissions'],
                 color='skyblue', alpha=0.4)

plt.plot(monthly_admissions['Date'],
         monthly_admissions['Admissions'],
         marker='o', color='blue', linewidth=2)

plt.title('Monthly Trends in Patient Admissions', fontsize=16)
plt.xlabel('Month')
plt.ylabel('Number of Admissions')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("Hospital  Health Care Management Data set.xlsx")

dept_patient_counts = (
    df.groupby('Dpt_ID')  
      .size()
      .reset_index(name='Patient_Count')
      .sort_values(by='Patient_Count', ascending=False)
)

location_patient_counts = (
    df.groupby('City') 
      .size()
      .reset_index(name='Patient_Count')
      .sort_values(by='Patient_Count', ascending=False)
)

plt.figure(figsize=(10, 6))
plt.bar(dept_patient_counts['Dpt_ID'],
        dept_patient_counts['Patient_Count'],
        color='skyblue', alpha=0.8)

plt.title('Patient Volume by Department', fontsize=16)
plt.xlabel('Department ID')
plt.ylabel('Number of Patients')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
plt.bar(location_patient_counts['City'],
        location_patient_counts['Patient_Count'],
        color='lightgreen', alpha=0.8)

plt.title('Patient Volume by City', fontsize=16)
plt.xlabel('City')
plt.ylabel('Number of Patients')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("Hospital  Health Care Management Data set.xlsx")

# Ensure LOS is numeric
df['LOS'] = pd.to_numeric(df['LOS'], errors='coerce')

bins = [0, 1, 3, 7, 14, 30, df['LOS'].max()]
labels = ['0-1 days', '2-3 days', '4-7 days', '8-14 days', '15-30 days', '30+ days']

df['LOS_Category'] = pd.cut(df['LOS'], bins=bins, labels=labels, right=True)

# Count patients per category
los_counts = df['LOS_Category'].value_counts().sort_index()
fig, ax = plt.subplots(figsize=(8, 8))
wedges, texts, autotexts = ax.pie(
    los_counts,
    labels=los_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    wedgeprops=dict(width=0.3),  # This makes it a donut
    colors=plt.cm.Paired.colors
)

# Center text
plt.setp(autotexts, size=10, weight="bold", color="white")
ax.set_title("Length of Stay (LOS) Distribution", fontsize=16)

# Add center label
plt.text(0, 0, "LOS", ha='center', va='center', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_excel("Hospital  Health Care Management Data set.xlsx")

df['treatemencost'] = pd.to_numeric(df['treatemencost'], errors='coerce')

treatment_stats = df.groupby('Patient type').agg(
    Patient_Volume=('Patient type', 'count'),
    Total_Treatment_Cost=('treatemencost', 'sum')
).reset_index()

x = np.arange(len(treatment_stats['Patient type']))  # positions
width = 0.4
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(x - width/2, treatment_stats['Patient_Volume'], width, label='Patient Volume', color='skyblue')
ax.bar(x + width/2, treatment_stats['Total_Treatment_Cost'], width, label='Total Treatment Cost', color='orange')

# Labels & formatting
ax.set_xlabel('Treatment Type')
ax.set_ylabel('Count / Cost')
ax.set_title('Treatment Type Breakdown: Patient Volume vs Total Cost')
ax.set_xticks(x)
ax.set_xticklabels(treatment_stats['Patient type'], rotation=45, ha='right')
ax.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel("Hospital  Health Care Management Data set.xlsx")
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Bed'] = pd.to_numeric(df['Bed'], errors='coerce')
df['Month_Year'] = df['Date'].dt.to_period('M')
bed_occupancy = df.groupby(['Dpt_ID', 'Month_Year'])['Bed'].count().reset_index()
heatmap_data = bed_occupancy.pivot(index='Dpt_ID', columns='Month_Year', values='Bed').fillna(0)

plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, cmap="YlOrRd", annot=True, fmt=".0f")

plt.title("Bed Occupancy by Department and Month", fontsize=16)
plt.xlabel("Month-Year")
plt.ylabel("Department")
plt.tight_layout()
plt.show()

```
